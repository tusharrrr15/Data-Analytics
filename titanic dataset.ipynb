```python
#drop duplicates
titanic.drop_duplicates(inplace = True)
titanic.reset_index(inplace = True, drop = True)
titanic
```

```python
#checing for null values
null_counts = titanic.isnull().sum()
if null_counts.sum()>0:
    print(null_counts[null_counts>0])
```

```python
#Data cleaning and missing value handling

#if number od entry is N, you have 5-10% missing value-- we can eliminate those rows
titanic = titanic[titanic['Embarked'].notna()]

#if the number of entry N,you have 30-50% miss value --we can replace the rows with mean,median,mode
#titanic['Age'].fillna(titanic['Age'].median(),inplace = True)

#if the missing values 70% above -- use some intution and eliminate the column
titanic = titanic.drop('Cabin', axis =1)
```

```python
titanic = titanic[titanic.Fare<500]
```

```python
import numpy as np
z_scores = np.abs((titanic[['Age','SibSp']]-titanic[['Age','SibSp']].mean()/titanic[['Age','SibSp']].std()))

#ZScore
outlier_threshold = 3

outliers = titanic[(z_scores['Age']>outlier_threshold) | (z_scores['SibSp']>outlier_threshold)]

print(outliers)
```

```python
#winsorize the outliers by replacing them with the 95th percentile value
age_threshold = titanic['Age'].quantile(0.95)
SibSp_threshold = titanic['SibSp'].quantile(0.95)

titanic['Age'] = np.where(titanic['Age']>age_threshold, age_threshold, titanic['Age'])
titanic['SibSp'] = np.where(titanic['SibSp']>SibSp_threshold,  SibSp_threshold, titanic['SibSp'])
```

```python
#drop columns which are not neccessary
titanic.drop(['PassengerId','Name','Ticket'],axis = 1, inplace=True)
titanic.reset_index(inplace=True, drop=True)
titanic
```

```python
#group by
survived = titanic.groupby('Survived').mean()
survived
```
